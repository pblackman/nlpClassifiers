```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
```

```python
import os
import logging
import gensim
from gensim.models import word2vec
from gensim.models.callbacks import CallbackAny2Vec
```

```python
os.environ["CUDA_VISIBLE_DEVICES"]="2"
```

```python
VOCABULARY_FILE = '../../data/twitter-en/vocab/tweets_EN_Corpus.csv'
EMBEDDING_DIM = 300
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
sentences = word2vec.LineSentence(VOCABULARY_FILE)
model = word2vec.Word2Vec(
    sentences,
    size=EMBEDDING_DIM,
    window=10,
    min_count=3,
    workers=10,
    iter=10)
```

```python
model.wv.most_similar_cosmul(positive=['volume'])
```

```python
model.wv.save_word2vec_format('../../data/twitter-en/vocab/word2vec-model-tweets-en.bin')
```
