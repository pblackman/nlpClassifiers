```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python
data = pd.read_csv("../../data/ml-best.csv", sep=";",names =['accuracy','support'], header=None) 
```

```python
data = data.sort_values(by=['support'])
```

```python
#plt.bar(data['support'],data['accuracy'])
#plt.legend({'Mean':mean})
f, ax = plt.subplots(figsize=(10,6)) # set the size that you'd like (width, height)
plt.scatter(data['support'],data['accuracy'],s=4)
#plt.bar([10,11,12,13], [0.4,0.3,0.2,0.1], label = 'second bar')
ax.legend(fontsize = 14)
plt.xlabel('Support')
plt.ylabel('Accuracy')
plt.title('Mercado Livre')
plt.show()
```
